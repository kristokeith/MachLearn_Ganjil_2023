1. Import

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

2. Load Data

```python
data = pd.read_csv('insurance.csv')
# inspect first data
data.head()

# check data size
data.shape

# check general information
data.info()

# check data description
data.describe()
```

3. Identify the variables that will be used as independent variables (features) and the target variable (personal medical costs).

```python
# visualize the data using pairplot
import matplotlib.pyplot as plt
import seaborn as sns
sns.pairplot(data, x_vars=['age', 'bmi', 'children'],
             y_vars='charges', size=4, aspect=1, kind='scatter')
plt.show()
# visualize the data correlation using heatmaps
sns.heatmap(data.corr(), cmap="YlGnBu", annot=True)
plt.show()
```

```python
X = data[['age']]  # Independent variables (features)
y = data['charges']  # Target variable
```

4. Split the dataset into training (train) and testing (test) data with an appropriate proportion.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, test_size=0.3, random_state=100)
```

5. Perform feature scaling if necessary. I useStandardization (z-score scaling) to scale the features.

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

6. Create a multiple linear regression model using Scikit-Learn.

```python
import statsmodels.api as sm

X_train_sm = sm.add_constant(X_train)
lr = sm.OLS(y_train, X_train_sm).fit()
plt.scatter(X_train, y_train)
plt.plot(X_train, 265.2483 + 66.3015*X_train, 'r')
plt.show()
```

7. Train the model on the training data and make predictions on the testing data.

```python
# Make prediction
y_train_pred = lr.predict(X_train_sm)

# Calculate residual
res = (y_train - y_train_pred)

# Histogram residual
fig = plt.figure()
sns.distplot(res, bins=15)
plt.title('........', fontsize=15)
plt.xlabel('y_train - y_train_pred', fontsize=15)
plt.show()

# Scatter plot residual
plt.scatter(X_train, res)
plt.show()
```

```python
# Make a prediction using testing data
X_test_sm = sm.add_constant(X_test)
y_test_pred = lr.predict(X_test_sm)

plt.scatter(X_test, y_test)
plt.plot(X_test, y_test_pred, 'r')
plt.show()
```

8. Evaluate the model by calculating metrics such as R-squared, MSE, and MAE. Display the evaluation results.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

svr = SVR(kernel='linear')
svr.fit(X_train_scaled, y_train)
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

y_pred = svr.predict(X_test_scaled)

r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

print("MAE:", mae)
print("MSE:", mse)
print("R-squared:", r2)
```
